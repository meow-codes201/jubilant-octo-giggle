```python
import numpy as np
import statistics
from scipy import stats
import math
import random

data = [5, 7, 3, 8, 10, 10, 3]
print("Mean (NumPy):", np.mean(data))
print("Mean (statistics):", statistics.mean(data))

print("Median (NumPy):", np.median(data))
print("Median (statistics):", statistics.median(data))

# Mode might return multiple in scipy
print("Mode (scipy):", stats.mode(data, keepdims=True).mode[0])
print("Mode (statistics):", statistics.mode(data))  # Only works if one mode exists
```

```python
#  BAYES THEOREM EXAMPLE

#     Prior: 1% of people have disease
#     Test is 99% accurate (true positive)
#     False positive rate = 5%
#     You test positive. Whats the probability you actually have the disease?

# Bayes formula
# P(D|T) = (P(T|D) * P(D)) / P(T)
# P(T) = P(T|D) * P(D) + P(T|D) * P(D)

p_d = 0.01             # Prior: has disease
p_not_d = 0.99         # Doesnt have disease

p_t_given_d = 0.99     # Test positive if diseased
p_t_given_not_d = 0.05 # False positive

p_t = p_t_given_d * p_d + p_t_given_not_d * p_not_d

p_d_given_t = (p_t_given_d * p_d) / p_t

print("P(Disease | Positive Test):", p_d_given_t)
```

```python
import random
import numpy as np

NUM_CITIES = 10
cities = np.random.rand(NUM_CITIES, 2) * 100

def total_distance(tour,cities):
  distance  = 0.0
  for i in range(len(tour)):
    from_city = tour[i]
    to_city = tour[(i+1) % len(tour)]
    distance += np.linalg.norm(cities[from_city] - cities[to_city])
  return distance

def create_population(pop_size,tour_size):
  population = []
  for i in range(pop_size):
    new_tour = list(range(tour_size))
    random.shuffle(new_tour)
    population.append(new_tour)
  return population
population = create_population(100,NUM_CITIES)
for i in range(3):
  print(population[i])
  print(total_distance(population[i],cities))
  print()
```

```python
# Check for names containing 'Neal' or 'Keith' (case-insensitive)
print("Names containing 'Neal' or 'Keith':")
display(bios[bios['name'].str.contains('Neal|Keith', case=False)].head())

# Check for names ending with 'sen' or 'san' (case-insensitive)
print("\nNames ending with 'sen' or 'san':")
display(bios[bios['name'].str.contains('sen$|san$', case=False)].head())

# Check for names starting with 'A' (case-sensitive)
print("\nNames starting with 'A':")
display(bios[bios['name'].str.contains('^A')].head())

# Check for names containing exactly 'Bob' (case-sensitive)
print("\nNames containing exactly 'Bob':")
display(bios[bios['name'].str.contains('^Bob$')].head())

# Check for names containing 'ille' or 'ollie' and ignore NaN values
print("\nNames containing 'ille' or 'ollie' (ignoring NaN):")
display(bios[bios['name'].str.contains('ille|ollie', na=False)].head())
```

Here's a breakdown of the different examples using `.str.contains()`:

1.  **`bios['name'].str.contains('Neal|Keith', case=False)`**: This checks if the 'name' column contains either "Neal" or "Keith". `case=False` makes the search case-insensitive.
2.  **`bios['name'].str.contains('sen$|san$', case=False)`**: This checks if the 'name' column contains strings that *end* with "sen" or "san". The `$` symbol in the pattern signifies the end of the string. `case=False` makes the search case-insensitive.
3.  **`bios['name'].str.contains('^A')`**: This checks if the 'name' column contains strings that *start* with "A". The `^` symbol in the pattern signifies the beginning of the string. By default, `case` is `True`, so this is a case-sensitive search.
4.  **`bios['name'].str.contains('^Bob$')`**: This checks if the 'name' column contains strings that are *exactly* "Bob". The `^` and `$` together match the entire string. By default, `case` is `True`, so this is a case-sensitive search.
5.  **`bios['name'].str.contains('ille|ollie', na=False)`**: This checks if the 'name' column contains either "ille" or "ollie". `na=False` ensures that any rows with missing values (NaN) in the 'name' column are excluded from the results.
